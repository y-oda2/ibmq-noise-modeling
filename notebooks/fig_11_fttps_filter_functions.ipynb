```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
color_sX = 'royalblue'
color_X = 'navy'
color_I = 'k'
color_CR = 'orange'
```

```python
import matplotlib as mpl
mpl.rc('text', usetex = True)
mpl.rc('font', family = 'serif')
```

```python
def gaussian(x, mu, sig):
    gg = np.exp(-np.power(x - mu, 2.) / (2 * np.power(sig, 2.))) - np.exp(-np.power(mu, 2.) / (2 * np.power(sig, 2.)))
    gg /= np.max(gg)
    return gg
```

## SPAM

```python
plt.figure(figsize=(10,1.5), dpi=150)

num_gates = 2

n = 100
nn = np.arange(n)
gg_X = gaussian(np.array(nn), nn[-1]/2, nn[-1]/4)
gg_sX = gg_X/2

plt.plot(nn, gg_X, color=color_X, lw=2)
plt.fill_between(nn,0,gg_X, alpha=0.5, color=color_X)
# plt.text(n/3,-0.55,'$X$', fontsize=20, color=color_X)

N = 1000 - num_gates*n
NN = np.arange(N)

xx = n+1 + np.array(NN)
yy = np.zeros(len(xx))
plt.plot(xx, yy, color='k', lw=0)

plt.ylim(-1.05,1.05)
plt.axis('off')
# plt.savefig('figures/exp_S.png')
plt.show()
```
